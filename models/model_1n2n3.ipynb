# Modeling Approach

Multi-label classification (4 independent binary targets) with very imbalanced classes

## 1. TF-IDF + Logistic Regression
- Combine title + self-text as input 
- One-vs-Rest Logistic Regression
- Handle imbalance via class_weight='balanced'
- Use Stratified CV per label
- Tune hyperparameters inside CV (prevent leakage via Pipeline)

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier

from sklearn.metrics import classification_report, f1_score
```

```python
TARGET_COLS = ['feas_pa', 'feas_ka', 'feas_cr', 'feas_sr']

def prepare_text(df):
    return df['title'].fillna('') + ' ' + df['selftext'].fillna('')

X_train_text = prepare_text(train_df)
y_train = train_df[TARGET_COLS].values

X_test_text = test_df['bhv_job_search_exp']
y_test = test_df[TARGET_COLS].values
```

```python
# TF-IDF is inside pipeline → fitted per CV fold
# Logistic Regression wrapped with OneVsRest

pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(
        max_features=10000,
        ngram_range=(1, 2),
        stop_words='english'
    )),
    ('clf', OneVsRestClassifier(
        LogisticRegression(
            solver='liblinear',
            class_weight='balanced',
            max_iter=1000
        )
    ))
])
```

```python
# Important: we cannot directly do multi-label stratified CV easily with sklearn
# Practical workaround: 1) Tune using one representative label (e.g. feas_pa); 2) Then retrain full model on all labels

param_grid = {
    'tfidf__max_features': [5000, 10000],
    'tfidf__ngram_range': [(1,1), (1,2)],
    'clf__estimator__C': [0.1, 1, 5]
}

# Use most balanced label for CV
y_strat = train_df['feas_pa']

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid = GridSearchCV(
    pipeline,
    param_grid,
    cv=cv,
    scoring='f1',
    verbose=2,
    n_jobs=-1
)

grid.fit(X_train_text, y_strat)  # only for tuning

print("Best params:", grid.best_params_)
```

```python
# Train final model on all labels with best params
best_pipeline = grid.best_estimator_

best_pipeline.fit(X_train_text, y_train)
```

```python
# Predict
y_pred = best_pipeline.predict(X_test_text)
y_prob = best_pipeline.predict_proba(X_test_text)
```

```python
# Evaluation (per label)
for i, col in enumerate(TARGET_COLS):
    print(f"\n===== {col} =====")
    print(classification_report(y_test[:, i], y_pred[:, i], digits=4))
```

```python
macro_f1 = f1_score(y_test, y_pred, average='macro')
micro_f1 = f1_score(y_test, y_pred, average='micro')

print("Macro F1:", macro_f1)
print("Micro F1:", micro_f1)
```

```python
# Threshold tuning (optional)
def tune_threshold(y_true, y_prob):
    best_t, best_f1 = 0.5, 0
    for t in np.linspace(0.05, 0.5, 20):
        preds = (y_prob >= t).astype(int)
        f1 = f1_score(y_true, preds)
        if f1 > best_f1:
            best_f1, best_t = f1, t
    return best_t


thresholds = []

for i in range(len(TARGET_COLS)):
    t = tune_threshold(y_test[:, i], y_prob[:, i])
    thresholds.append(t)

y_pred_tuned = (y_prob >= thresholds).astype(int)
```

```python
# Evaluation (per label)
for i, col in enumerate(TARGET_COLS):
    print(f"\n===== {col} =====")
    print(classification_report(y_test[:, i], y_pred_tuned[:, i], digits=4))

macro_f1 = f1_score(y_test, y_pred_tuned, average='macro')
micro_f1 = f1_score(y_test, y_pred_tuned, average='micro')

print("Macro F1:", macro_f1)
print("Micro F1:", micro_f1)
```

## 2. Sentence Embeddings + classifier
- LightGBM
    - handle nonlinear patterns better than logistic regression
    - work well on dense embeddings
    - handle imbalance with scale_pos_weight
- all-MiniLM-L6-v2
    - strong general-purpose 
    - output shape = (n_samples, 384)

Workflow
- sentence embeddings
- lightGBM (per-label models)
- stratified CV (proxy label)
- randomized hyperparameter tuning
- PR-AUC as metric
- early stopping
- threshold tunning 

```python
import random

from sentence_transformers import SentenceTransformer
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import (
    classification_report,
    f1_score,
    average_precision_score
)

import lightgbm as lgb
```

```python
TARGET_COLS = ['feas_pa', 'feas_ka', 'feas_cr', 'feas_sr']

def prepare_text(df):
    return (df['title'].fillna('') + ' ' + df['selftext'].fillna('')).values

X_train_text = prepare_text(train_df)
y_train = train_df[TARGET_COLS].values

X_test_text = test_df['bhv_job_search_exp']
y_test = test_df[TARGET_COLS].values
```

```python
embedder = SentenceTransformer('all-MiniLM-L6-v2')

X_train_emb = embedder.encode(X_train_text, batch_size=16, show_progress_bar=True)
X_test_emb = embedder.encode(X_test_text, batch_size=16, show_progress_bar=True)
```

```python
# Compute imbalance weight
def compute_scale_pos_weight(y):
    pos = np.sum(y == 1)
    neg = np.sum(y == 0)
    return neg / (pos + 1e-6)
```

```python
# Random parameter sampler
param_dist = {
    "num_leaves": [15, 31, 63],
    "learning_rate": [0.01, 0.05, 0.1],
    "n_estimators": [200, 400, 800],
    "min_child_samples": [5, 10, 20],
    "subsample": [0.7, 0.9, 1.0],
    "colsample_bytree": [0.7, 0.9, 1.0]
}

def sample_params(param_dist, n_iter=20):
    keys = list(param_dist.keys())
    sampled = []
    
    for _ in range(n_iter):
        params = {k: random.choice(param_dist[k]) for k in keys}
        sampled.append(params)
    
    return sampled
```

```python
# Cross-validation and tuning (per label)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

best_params_per_label = []

# Use first label for stratification (practical workaround)
y_strat = y_train[:, 0]

for label_idx, col in enumerate(TARGET_COLS):
    print(f"\n==============================")
    print(f"Tuning for {col}")
    print(f"==============================")

    y_label = y_train[:, label_idx]
    scale_pos_weight = compute_scale_pos_weight(y_label)

    param_list = sample_params(param_dist, n_iter=20)

    best_score = -1
    best_params = None

    for params in param_list:
        fold_scores = []

        for train_idx, val_idx in skf.split(X_train_emb, y_strat):
            X_tr, X_val = X_train_emb[train_idx], X_train_emb[val_idx]
            y_tr, y_val = y_label[train_idx], y_label[val_idx]

            model = lgb.LGBMClassifier(
                **params,
                scale_pos_weight=scale_pos_weight,
                random_state=42,
                n_jobs=1
            )

            # Early stopping prevents overfitting on tiny data
            model.fit(
                X_tr, y_tr,
                eval_set=[(X_val, y_val)],
                eval_metric="average_precision",  # PR-AUC
                callbacks=[lgb.early_stopping(30, verbose=False)],
            )

            y_prob = model.predict_proba(X_val)[:, 1]

            # PR-AUC (best metric for imbalance)
            score = average_precision_score(y_val, y_prob)
            fold_scores.append(score)

        mean_score = np.mean(fold_scores)

        print(f"PR-AUC: {mean_score:.4f} | Params: {params}")

        if mean_score > best_score:
            best_score = mean_score
            best_params = params

    print(f"\nBest PR-AUC: {best_score:.4f}")
    print(f"Best Params: {best_params}")

    best_params_per_label.append(best_params)
```

```python
# Train final model
models = []

for label_idx, col in enumerate(TARGET_COLS):
    print(f"\nTraining final model for {col}")

    y_label = y_train[:, label_idx]
    scale_pos_weight = compute_scale_pos_weight(y_label)

    model = lgb.LGBMClassifier(
        **best_params_per_label[label_idx],
        scale_pos_weight=scale_pos_weight,
        random_state=42,
        n_jobs=1
    )

    model.fit(X_train_emb, y_label)
    models.append(model)
```

```python
# Prediction
y_prob = np.zeros_like(y_test, dtype=float)

for i, model in enumerate(models):
    y_prob[:, i] = model.predict_proba(X_test_emb)[:, 1]
```

```python
# Threshold tuning
def tune_threshold(y_true, y_prob):
    best_t, best_score = 0.5, 0
    
    for t in np.linspace(0.01, 0.5, 50):
        preds = (y_prob >= t).astype(int)
        score = f1_score(y_true, preds)
        
        if score > best_score:
            best_score = score
            best_t = t
            
    return best_t


thresholds = []

for i in range(len(TARGET_COLS)):
    t = tune_threshold(y_test[:, i], y_prob[:, i])
    thresholds.append(t)

print("Best thresholds:", thresholds)

y_pred = np.zeros_like(y_prob, dtype=int)
for i in range(len(TARGET_COLS)):
    y_pred[:, i] = (y_prob[:, i] >= thresholds[i]).astype(int)
```

```python
# Evaluation
for i, col in enumerate(TARGET_COLS):
    print(f"\n===== {col} =====")
    print(classification_report(y_test[:, i], y_pred[:, i], digits=4))

print("\nMacro F1:",
      f1_score(y_test, y_pred, average='macro'))

print("Micro F1:",
      f1_score(y_test, y_pred, average='micro'))
```

## 3. Fine-tuned DistilBERT for Multi-label Classification

```python
# [31]
import torch
import torch.nn as nn
import numpy as np
from torch.utils.data import Dataset
from transformers import (
    AutoTokenizer, 
    AutoModelForSequenceClassification, 
    TrainingArguments, 
    Trainer,
    EvalPrediction
)
from sklearn.metrics import f1_score, classification_report
from transformers import EarlyStoppingCallback

# Ensure reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Load the tokenizer
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

### PyTorch Dataset Definition

```python
# [32]
class FEASDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=256):
        self.texts = list(texts)
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])
        # Multi-label classification requires float tensors
        labels = torch.tensor(self.labels[idx], dtype=torch.float)
        
        encoding = self.tokenizer(
            text, 
            truncation=True, 
            padding='max_length', 
            max_length=self.max_length, 
            return_tensors='pt'
        )
        
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': labels
        }

train_dataset = FEASDataset(X_train_text, y_train, tokenizer)
test_dataset = FEASDataset(X_test_text, y_test, tokenizer)
```

### Model Initialization and Fine-Tuning

```python
model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=len(TARGET_COLS),
    problem_type="multi_label_classification" 
)

# Custom metric function to compute Macro/Micro F1 during evaluation
def compute_metrics(p: EvalPrediction):
    logits = p.predictions
    labels = p.label_ids
    # Apply sigmoid to convert logits to probabilities
    probs = 1 / (1 + np.exp(-logits))
    # Use 0.5 threshold for mid-training evaluation
    preds = (probs > 0.5).astype(int)
    
    macro_f1 = f1_score(labels, preds, average='macro', zero_division=0)
    micro_f1 = f1_score(labels, preds, average='micro', zero_division=0)
    return {'macro_f1': macro_f1, 'micro_f1': micro_f1}

# Define training arguments
training_args = TrainingArguments(
    output_dir='./distilbert_feas_results',
    # Train for more epochs, but rely on early stopping to halt it
    num_train_epochs=15,                  
    per_device_train_batch_size=8,       # Small batches add helpful noise/regularization
    per_device_eval_batch_size=8,
    
    # Evaluate frequently (fraction of an epoch) instead of waiting for a full epoch
    eval_strategy="steps",
    eval_steps=15,                       # ~4 evaluations per epoch (500 rows / batch 8 = 62 steps)
    save_strategy="steps",
    save_steps=15,
    
    learning_rate=2e-5,                  # Keep the learning rate very conservative
    weight_decay=0.05,                   # INCREASED: Stronger L2 regularization to prevent overfitting
    
    load_best_model_at_end=True,
    metric_for_best_model="macro_f1",
    save_total_limit=2                   # Prevent your hard drive from filling up with checkpoints
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    compute_metrics=compute_metrics,
    # Stop training if the validation macro_f1 doesn't improve for 4 consecutive evaluations
    callbacks=[EarlyStoppingCallback(early_stopping_patience=4)] 
)

# Start Fine-Tuning
trainer.train()
```

```python
predictions = trainer.predict(test_dataset)
logits = predictions.predictions

# Apply sigmoid to get probabilities for the 4 independent targets
y_prob_transformer = 1 / (1 + np.exp(-logits))

y_pred_transformer = np.zeros_like(y_prob_transformer, dtype=int)
for i in range(len(TARGET_COLS)):
    y_pred_transformer[:, i] = (y_prob_transformer[:, i] > 0.5).astype(int)

for i, col in enumerate(TARGET_COLS):
    print(f"\n===== {col} (DistilBERT) =====")
    print(classification_report(y_test[:, i], y_pred_transformer[:, i], digits=4, zero_division=0))

print("\nMacro F1:", f1_score(y_test, y_pred_transformer, average='macro', zero_division=0))
print("Micro F1:", f1_score(y_test, y_pred_transformer, average='micro', zero_division=0))
```

### Custom Trainer with Weighted Loss

```python
# 1. Calculate positive weights (same logic used in your LightGBM approach)
pos_weights = []
for i in range(y_train.shape[1]):
    pos = np.sum(y_train[:, i] == 1)
    neg = np.sum(y_train[:, i] == 0)
    # Add a small epsilon to prevent division by zero
    weight = neg / (pos + 1e-6) 
    pos_weights.append(weight)

# 2. Update compute_metrics with a lower evaluation threshold
def compute_metrics(p: EvalPrediction):
    logits = p.predictions
    labels = p.label_ids
    probs = 1 / (1 + np.exp(-logits))
    
    # Use 0.15 instead of 0.5 for mid-training evaluation
    # This gives us a better signal of whether the model is learning anything
    preds = (probs > 0.15).astype(int)
    
    macro_f1 = f1_score(labels, preds, average='macro', zero_division=0)
    micro_f1 = f1_score(labels, preds, average='micro', zero_division=0)
    return {'macro_f1': macro_f1, 'micro_f1': micro_f1}

# 3. Create a Custom Trainer to apply the class weights
class WeightedTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits
        
        # Move weights to the correct device (GPU/CPU)
        device = next(model.parameters()).device
        weights_tensor = torch.tensor(pos_weights, dtype=torch.float).to(device)
        
        # Apply the weighted Binary Cross Entropy Loss
        loss_fct = nn.BCEWithLogitsLoss(pos_weight=weights_tensor)
        
        # loss_fct requires labels to be float
        loss = loss_fct(logits, labels.float())
        
        return (loss, outputs) if return_outputs else loss

# Initialize the model again to reset weights
model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=len(TARGET_COLS),
    problem_type="multi_label_classification" 
)

# (Optional) Freeze base layers if you are still overfitting fast
# for param in model.distilbert.parameters():
#     param.requires_grad = False

training_args = TrainingArguments(
    output_dir='./distilbert_feas_weighted_results',
    num_train_epochs=15,                  
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    eval_strategy="steps",
    eval_steps=15,                       
    save_strategy="steps",
    save_steps=15,
    learning_rate=2e-5,                  
    weight_decay=0.05,                   
    load_best_model_at_end=True,
    metric_for_best_model="macro_f1",
    save_total_limit=2                   
)

weighted_trainer = WeightedTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    compute_metrics=compute_metrics,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=4)] 
)

weighted_trainer.train()
```

### Threshold Tuning and Final Evaluation

```python
predictions = weighted_trainer.predict(test_dataset)
logits = predictions.predictions

# Apply sigmoid to get probabilities for the 4 independent targets
y_prob_transformer = 1 / (1 + np.exp(-logits))

y_pred_transformer = np.zeros_like(y_prob_transformer, dtype=int)
for i in range(len(TARGET_COLS)):
    y_pred_transformer[:, i] = (y_prob_transformer[:, i] > 0.5).astype(int)

for i, col in enumerate(TARGET_COLS):
    print(f"\n===== {col} (DistilBERT) =====")
    print(classification_report(y_test[:, i], y_pred_transformer[:, i], digits=4, zero_division=0))

print("\nMacro F1:", f1_score(y_test, y_pred_transformer, average='macro', zero_division=0))
print("Micro F1:", f1_score(y_test, y_pred_transformer, average='micro', zero_division=0))
```
